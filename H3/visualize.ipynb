```python
import matplotlib.pyplot as plt
import scipy 
from scipy.io import loadmat

plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
    "font.serif": ["Palatino"],  # O Times, Garamond, etc.
})
from matplotlib.ticker import PercentFormatter

plt.rcParams['font.size'] = 14  # Tamaño de fuente general
plt.rcParams['axes.titlesize'] = 18  # Tamaño del título de los ejes
plt.rcParams['axes.labelsize'] = 14  # Tamaño de las etiquetas de los ejes (X e Y)
plt.rcParams['xtick.labelsize'] = 12  # Tamaño de las etiquetas de los ticks del eje X
plt.rcParams['ytick.labelsize'] = 12  # Tamaño de las etiquetas de los ticks del eje Y
plt.rcParams['legend.fontsize'] = 12  # Tamaño de la fuente de la leyenda
plt.rcParams['figure.titlesize'] = 18 # Tamaño del título de la figura (suptitle)
```

```python
data_ng = loadmat('./backup/saved.mat')
data_G = loadmat('./backup/saved_cluster.mat')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(8, 3))

axes[0].plot(data_ng['n_gs'][0], data_ng['values_to_show'][0]/100, color='k', marker='.')
axes[0].axvline(x=620, color='red', linestyle='--', linewidth=0.5, label=r'$N_g=620$')
axes[0].set_xlabel(r'Numero de individuos por grupo ($N_g$)')
axes[0].set_ylabel(r'Poder Empirico\\(Rechazo $H_0: \beta = 0$)')
valor_horizontal = data_ng['values_to_show'][0][3] 
axes[0].axhline(y=valor_horizontal/100, color='gray', linestyle='--', linewidth=0.8, label='Poder={:.1f}\%'.format(valor_horizontal))
axes[0].legend()
axes[0].set_xscale('log')
axes[0].yaxis.set_major_formatter(PercentFormatter(xmax=1.0))

axes[1].plot(data_G['Gs'][0], data_G['values_to_show'][0]/100, color='k', marker='.')
axes[1].axvline(x=4, color='red', linestyle='--', linewidth=0.5, label=r'$G=4$')
axes[1].set_xlabel(r'Numero de Grupos ($G$)')
# axes[1].set_ylabel(r'Poder Empirico\\(Rechazo $H_0: \beta = 0$)')
valor_horizontal = data_G['values_to_show'][0][0] 
axes[1].axhline(y=valor_horizontal/100, color='gray', linestyle='--', linewidth=0.8, label='Poder={:.1f}\%'.format(valor_horizontal))
axes[1].legend()
axes[1].set_xscale('log')
axes[1].yaxis.set_major_formatter(PercentFormatter(xmax=1.0))

fig.savefig('./figures/poder.pdf', dpi=300, bbox_inches='tight',)
```
